```python
#os.environ['KMP_DUPLICATE_LIB_OK']='TRUE'
#„Åø„Å™„Åø„ÇìÁå´„ÅÆÂà§ÂÆö
import keras
import sys, os
import numpy as np
from PIL import Image
from keras.models import load_model
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

imsize = (64, 64)

testpic     = "./IMG_1134.JPG"
keras_param = "./cnn.h5"

def load_image(path):
    img = Image.open(path)
    img = img.convert('RGB')
    # Â≠¶ÁøíÊôÇ„Å´„ÄÅ(64, 64, 3)„ÅßÂ≠¶Áøí„Åó„Åü„ÅÆ„Åß„ÄÅÁîªÂÉè„ÅÆÁ∏¶„ÉªÊ®™„ÅØ‰ªäÂõû Â§âÊï∞imsize„ÅÆ(64, 64)„Å´„É™„Çµ„Ç§„Ç∫„Åó„Åæ„Åô„ÄÇ
    img = img.resize(imsize)
    # ÁîªÂÉè„Éá„Éº„Çø„ÇínumpyÈÖçÂàó„ÅÆÂΩ¢Âºè„Å´Â§âÊõ¥
    img = np.asarray(img)
    img = img / 255.0
    return img

model = load_model(keras_param)
img = load_image(testpic)
prd = model.predict(np.array([img]))
print(prd) # Á≤æÂ∫¶„ÅÆË°®Á§∫
print(prd[0][0])
print(prd[0][1])
prelabel = np.argmax(prd, axis=1)

print("----------Âà§ÂÆö----------")
if prelabel == 0:
    print("-> Áä¨üêï")
elif prelabel == 1:
    print("-> Áå´üêà")
print("----------Ë©≥Á¥∞----------")
Per_Dog = prd[0][0] * 100
Per_Cat = prd[0][1] * 100
print("Áä¨üêï: " + f"{Per_Dog:.3f}" + "%")
print("Áå´üêà: " + f"{Per_Cat:.3f}" + "%")
print("----------ÁîªÂÉè----------")
pic_width = 300
img = Image.open(testpic)
img.resize(( int(pic_width*(img.width/img.height)),pic_width ))
```
